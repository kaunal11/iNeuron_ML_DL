# **Problem Statement 1:**
In each of the following situations, state whether it is a correctly stated hypothesis
testing problem and why?
1. 𝐻0: 𝜇 = 25, 𝐻1: 𝜇 ≠ 25
2. 𝐻0: 𝜎 > 10, 𝐻1: 𝜎 = 10
3. 𝐻0: 𝑥 = 50, 𝐻1: 𝑥 ≠ 50
4. 𝐻0: 𝑝 = 0.1, 𝐻1: 𝑝 = 0.5
5. 𝐻0: 𝑠 = 30, 𝐻1: 𝑠 > 30

# **SOLUTION**

1. 𝐻0: 𝜇 = 25, 𝐻1: 𝜇 ≠ 25: **This is correct hypothesis and it is a two tailed test(𝜇<25 or 𝜇>25)**

2. 𝐻0: 𝜎 > 10, 𝐻1: 𝜎 = 10: **This is incorrect hypothesis and H1 should be 𝜎 < 10**

3. 𝐻0: 𝑥 = 50, 𝐻1: 𝑥 ≠ 50: **This is correct hypothesis and it is a two tailed test (x>50 or x<50)**

4. 𝐻0: 𝑝 = 0.1, 𝐻1: 𝑝 = 0.5: **This is incorrect hypothesis and H1 should be p ≠ 0.1**

1. 5. 𝐻0: 𝑠 = 30, 𝐻1: 𝑠 > 30: **This is incorrect hypothesis and H1 should be s ≠ 30**

```python
def z_stat(sam_mean,pop_mean,se):
    return (sam_mean-pop_mean)/se
def t_stat(sam_mean,pop_mean,se):
    return (sam_mean-pop_mean)/se
def sample_error(std_dev,n):
    return std_dev/np.sqrt(n)
```

# **Problem Statement 2:**

The college bookstore tells prospective students that the average cost of its
textbooks is Rs. 52 with a standard deviation of Rs. 4.50. A group of smart statistics
students thinks that the average cost is higher. To test the bookstore’s claim against
their alternative, the students will select a random sample of size 100. Assume that
the mean from their random sample is Rs. 52.80. Perform a hypothesis test at the
5% level of significance and state your decision.

```python
# H0 or Null Hypothesis = Average cost of textbooks is Rs. 52
# H1 or Alternate Hypothesis = Average cost of textbooks is not equal to Rs. 52 (Two-tailed test)

alpha=0.05
pop_mean=52
std_dev=4.5
n=100
sam_mean=52.80
se=sample_error(std_dev,n)
z_test=z_stat(pop_mean,sam_mean,se)
z_alpha=st.norm.ppf(alpha/2)
# p=2*(1-(st.norm.cdf(np.absolute(z))))
if -abs(z_alpha)<z_test<abs(z_alpha): # z_test is not in the critical region at both ends
    print("Null Hypothesis is accepted, 𝜇=52")
else:
    print("Null Hypothesis is rejected,  𝜇!=52")
```

# **Problem Statement 3:**

A certain chemical pollutant in the Genesee River has been constant for several
years with mean μ = 34 ppm (parts per million) and standard deviation σ = 8 ppm. A
group of factory representatives whose companies discharge liquids into the river is
now claiming that they have lowered the average with improved filtration devices. A
group of environmentalists will test to see if this is true at the 1% level of
significance. Assume \ that their sample of size 50 gives a mean of 32.5 ppm.
Perform a hypothesis test at the 1% level of significance and state your decision.

```python
# H0 or Null Hypothesis = Average pollution level is 34 ppm
# H1 or Alternate Hypothesis = Average pollution level is less than 34 ppm (One-tailed test)

alpha=0.01
pop_mean=34
std_dev=8
n=50
sam_mean=32.5
se=sample_error(std_dev,n)
z_test=z_stat(pop_mean,sam_mean,se)
z_alpha=st.norm.ppf(alpha)
if z_test<z_alpha:
    print("Null Hypothesis is rejected, 𝜇<34")
else:
    print("Null Hypothesis is accepted,  𝜇=34")
```

# **Problem Statement 4:**

Based on population figures and other general information on the U.S. population,
suppose it has been estimated that, on average, a family of four in the U.S. spends
about $1135 annually on dental expenditures. Suppose further that a regional dental
association wants to test to determine if this figure is accurate for their area of
country. To test this, 22 families of 4 are randomly selected from the population in
that area of the country and a log is kept of the family’s dental expenditure for one
year. The resulting data are given below. Assuming, that dental expenditure is
normally distributed in the population, use the data and an alpha of 0.5 to test the
dental association’s hypothesis.
1008, 812, 1117, 1323, 1308, 1415, 831, 1021, 1287, 851, 930, 730, 699,
872, 913, 944, 954, 987, 1695, 995, 1003, 994

```python
# H0 or Null Hypothesis = Average dental expense is $1135
# H1 or Alternate Hypothesis = Average dental expense is not $1135 (Two-tailed test)

alpha=0.5
pop_mean=1135
expense=[1008, 812, 1117, 1323, 1308, 1415, 831, 1021, 1287, 851, 930, 730, 699, 872, 913, 944,
         954, 987, 1695, 995, 1003, 994]
std_dev=np.std(expense)
n=22
sam_mean=np.mean(expense)
se=sample_error(std_dev,n)
t_test=t_stat(pop_mean,sam_mean,se)
t_alpha=st.t.ppf(1-alpha/2,n)
if -abs(t_alpha)<t_test<abs(t_alpha):# t_test is not in the critical region at both ends
    print("Null Hypothesis is accepted, 𝜇=$1135")
else:
    print("Null Hypothesis is rejected,  𝜇!=$1135")
```

# **Problem Statement 11:**
Two-tailed test for difference between two population means
Is there evidence to conclude that the number of people travelling from Bangalore to
Chennai is different from the number of people travelling from Bangalore to Hosur in
a week, given the following:
Population 1: Bangalore to Chennai n1 = 1200,x1 = 452,s1 = 212
Population 2: Bangalore to Hosur n2 = 800,x2 = 523,s2 = 185

```python
# H0 or Null Hypothesis = No. of people travelling from Bangalore to Chennai is same as the No. of people travelling from Bangalore to Hosur in a week
# H1 or Alternate Hypothesis = No. of people travelling from Bangalore to Chennai is different than the No. of people travelling from Bangalore to Hosur in a week (Two-tailed test)

n1=1200 # Size of Sample 1
x1=452 # Mean of Sample 1
s1=212 # Standard deviation of Sample 1

n2=800 # Size of Sample 2
x2=523 # Mean of Sample 2
s2=185 # Standard deviation of sample 2

alpha=0.05

std_sample=np.sqrt((s1**2+s2**2)/(n1+n2-2))
se=np.sqrt((std_sample/n1)+(std_sample/n2))

z_test=(x1-x2)/se
z_alpha=st.norm.ppf(alpha/2)

if -abs(z_alpha)<z_test<abs(z_alpha): # z_test is not in the critical region at both ends
    print("Null Hypothesis is accepted, 𝜇=Number of people traveling is same")
else:
    print("Null Hypothesis is rejected,  𝜇!=Number of people traveling is different")
```

# **Problem Statement 17:**
A die is thrown 132 times with the following results: Number turned up: 1, 2, 3, 4, 5, 6
Frequency: 16, 20, 25, 14, 29, 28
Is the die unbiased? Consider the degrees of freedom as 𝑝 − 1.

```python
# Null Hypothesis (H0): The dice is unbiased
# Alternate Hypothesis (H1): The dice is biased

observed=np.array([16,20,25,14,29,28])
expected=np.array([22,22,22,22,22,22])

chi,p=st.chisquare(f_obs=observed,f_exp=expected)
if p<0.05:
    print("Null Hypothesis is rejected i.e. the dice is biased")
else:
    print("Null Hypothesis is accepted i.e. the dice is unbiased")
```

# **Problem Statement 18:**

In a certain town, there are about one million eligible voters. A simple random sample of 10,000 eligible voters was chosen to study the relationship between gender and participation in the last election. The results are summarized in the following 2X2 (read two by two) contingency table: Men Women Total Voted 2792 3591 6383 Not Voted 1486 2131 3617 Total 4278 5722 10000 We would want to check whether being a man or a woman (columns) is independent of having voted in the last election (rows). In other words, is “gender and voting independent”?

```python
# Null Hypothesis H0 :Gender and voting is independent 
# Alternative Hypothesis H1 : Gender and voting are dependent

alpha=0.05
inp=np.array([[2792,3591],[1486,2131]])
chi2,p,dof,expected=st.chi2_contingency(inp)
print("The expected values are :",expected)
if p<alpha:
    print("Null Hypothesis is rejected(Gender and voting are dependent)")
else:
    print("Null Hypothesis is accepted(Gender and voting is independent)")
```
