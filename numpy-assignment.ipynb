# **Write a function so that the columns of the output matrix are powers of the input vector. The order of the powers is determined by the increasing boolean argument. Specifically, when increasing is False, the i-th output column is the input vector raised element-wise to the power of N - i - 1. HINT: Such a matrix with a geometric progression in each row is named for Alexandre-Theophile Vandermonde.** #

```python
def gen_vander_matrix(input_vector,n,increasing=False):
    
    if not increasing:
        output_vector = np.array([x**(n-1-i) for x in input_vector for i in range(n)]).reshape(input_vector.size,n)
    elif increasing:
        output_vector = np.array([x**i for x in input_vector for i in range(n)]).reshape(input_vector.size,n)
    
    return output_vector

print("------------OUTPUT-------------\n")

input_array = np.array([1,2,3,4,5])
n = 3
op_matx_dec_order = gen_vander_matrix(input_array,n,False)
op_matx_inc_order = gen_vander_matrix(input_array,n,True)

print("The input array is:",input_array,"\n")
print("Number of columns in output matrix should be:",n,"\n")
print("Vander matrix of the input array in decreasing order of powers:\n\n",op_matx_dec_order,"\n")
print("Vander matrix of the input array in increasing order of powers:\n\n",op_matx_inc_order,"\n")

inputvector = np.array([1,2,4,6,8,10])
no_col_opmat = 5
op_matx_dec_order = gen_vander_matrix(inputvector,no_col_opmat,False)
op_matx_inc_order = gen_vander_matrix(inputvector,no_col_opmat,True)

print("---------------------------------------------------------------\n")
print("The input array is:",inputvector,"\n")
print("Number of columns in output matrix should be:",no_col_opmat,"\n")
print("Vander matrix of the input array in decreasing order of powers:\n\n",op_matx_dec_order,"\n")
print("Vander matrix of the input array in increasing order of powers:\n\n",op_matx_inc_order,"\n")
```

# **Write a function to find moving average in an array over a window:Test it over [3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150] and window of 3.**#

```python
def moving_avg(arr,k):
    length=len(arr)-k+1
    new_arr=[]
    for i in range(length):
        new_arr.append(sum(arr[i:i+3])/k)
    return new_arr

my_array=[3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150]
k=3

moving_avg(my_array,k)
```
