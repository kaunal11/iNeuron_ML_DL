# **1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()**

```python
# We will build our own Reduce function to calculate the sum of numbers in a list

def add(a, b): 
    return a + b

def myreduce(func, seq):
    first = seq[0]
    for i in seq[1:]:
        first = func(first, i)
    return first
myreduce(add, [1, 2, 3, 4,5])
```

# ** 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()**

```python
# We will build our own Filter function to check which of the numbers in a list are even numbers

def even_check(num):
    if num%2 ==0:
        return True
    
def myfilter(seq):
    even=[s for s in seq if even_check(s)]
    
    return even

myfilter([1,2])
```
