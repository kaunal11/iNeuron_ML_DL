```python
import numpy as np # linear algebra
import scipy.stats as st
```

```python
def z_stat(sam_mean,pop_mean,se):
    return (sam_mean-pop_mean)/se
def t_stat(sam_mean,pop_mean,se):
    return (sam_mean-pop_mean)/se
def sample_error(std_dev,n):
    return std_dev/np.sqrt(n)
```

# **Problem Statement 1:**
Blood glucose levels for obese patients have a mean of 100 with a standard deviation of 15. A researcher thinks that a diet high in raw cornstarch will have a positive effect on
blood glucose levels. A sample of 36 patients who have tried the raw cornstarch diet
have a mean glucose level of 108. Test the hypothesis that the raw cornstarch had an
effect or not.

```python
# Null Hypothesis (H0): A diet high in raw cornstarch has no effect on blood glucose levels i.e. Mean =100
# Alternate Hypothesis (H1): A diet high in raw cornstarch has significant effect on blood glucose levels i.e. Mean!=100 (Two-tailed test)

n=36
std_dev=15
pop_mean=100
sam_mean=108
alpha=0.05

se=sample_error(std_dev,n)
z_test=z_stat(pop_mean,sam_mean,se)
# z_alpha=st.norm.ppf(alpha/2)
p=2*(1-(st.norm.cdf(np.absolute(z_test))))
if p<alpha: 
    print("Null Hypothesis is rejected, Mean Blood glucose levels! =100 ")
else:
    print("Null Hypothesis is accepted,  Mean Blod Glucose levels =100")
```

# **Problem Statement 3:**
You take the SAT and score 1100. The mean score for the SAT is 1026 and the standard
deviation is 209. How well did you score on the test compared to the average test taker?

```python
my_score=1100
mean_score=1026
std_dev=209
z=(my_score-mean_score)/std_dev
print("I score more than {percent} % of the students".format(percent='%.2f'%(st.norm.cdf(z)*100)))
```
